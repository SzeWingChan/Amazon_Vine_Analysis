```python
# Add the Pandas dependency
import pandas as pd
```

```python
# File to load
data = "Resources/vine_table.csv"

# Read the data file and store it in a Pandas DataFrame
df = pd.read_csv(data)
df
```

```python
# Filter the data and create a new DataFrame where the total_votes count 
# is equal to or greater than 20 to pick reviews 
vine_tv_df = df[(df["total_votes"] >= 20)]
vine_tv_df
```

```python
# Filter the data and create a new DataFrame to retrieve all the rows where the number of helpful_votes divided by 
# total_votes is equal to or greater than 50%

filtered_df = vine_tv_df[(vine_tv_df["helpful_votes"]/vine_tv_df["total_votes"] >=0.5)]
filtered_df
```

```python
# Filter and create a new DataFrame that retrieves all the rows where a review was written as part of the Vine program
vine_df = filtered_df[(filtered_df["vine"]=="Y")]
vine_df
```

```python
# Filter and create a new DataFrame that retrieves all the rows where the review was not 
# part of the Vine program (unpaid)
not_vine_df = filtered_df[(filtered_df["vine"]=="N")]
not_vine_df
```

```python
# Total number of reviews of the Vine program
vine_review_count = vine_df["review_id"].count()
vine_review_count
```

```python
# Total number of reviews that are not part of the Vine program
not_vine_review_count = not_vine_df["review_id"].count()
not_vine_review_count
```

```python
# The number of 5-star reviews of the Vine program
vine_5starreview_count = vine_df["star_rating"].loc[(vine_df["star_rating"]==5)].count()
vine_5starreview_count
```

```python
# The number of 5-star reviews that are not part of the Vine program
not_vine_5starreview_count = not_vine_df["star_rating"].loc[(not_vine_df["star_rating"]==5)].count()
not_vine_5starreview_count
```

```python
# The percentage of 5-star reviews of the Vine program
vine_review_p = vine_5starreview_count / vine_review_count * 100
vine_review_p 
```

```python
# The number of 5-star reviews for not part of the Vine program
not_vine_review_p = not_vine_5starreview_count / not_vine_review_count * 100
not_vine_review_p
```
